```python
from sklearn.datasets import load_iris
                                                                
iris=load_iris()                                                   # Loading data
                                                               
x=iris.data                                                        # Create feature and target arrays
y=iris.target 


                                                                   # Split into training and test set with 30% testdata 
from sklearn.model_selection import train_test_split     
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)


print("\n1. Extracting Feature :: \n")
print(x_test)

print("\n2. No:of rows and columns of test data :: ", end=" ")
print(x_test.shape)

print("\n3. No:of rows and columns of train data :: ", end=" ")
print(x_train.shape)

print("\n4. Different values test data :: ", end=" ")
print(y_test.shape)

print("\n5. Different values train data :: ", end=" ")
print(y_train.shape)


from sklearn.neighbors import KNeighborsClassifier

c_knn=KNeighborsClassifier(n_neighbors=3)
c_knn.fit(x_train,y_train)
y_pred=c_knn.predict(x_test)
print("\n6. Predicted output of test data :: ", end=" ")
print(y_pred)


from sklearn import metrics
print("\n7. Accuracy ::", metrics.accuracy_score(y_test,y_pred))

sample=[[5,5,3,2]]
pred=c_knn.predict(sample)
pred_v=[iris.target_names[p] for p in pred]
print("\n8. Class or predicted values ::",pred_v)
```
